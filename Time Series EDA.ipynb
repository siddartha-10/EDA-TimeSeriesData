## Using loc propetry
    The loc property is used to access a group of rows and columns by label(s) or a boolean array

```python
index = df_tesla.loc['2020-01-01':'2021-09-01'].index
stock_open = df_tesla.loc['2020-01-01':'2021-09-01']['Open']
```

## SUBPLOTS

```python
figure,axis = plt.subplots()
plt.tight_layout()
## Preventing overlapping od the dates on the x-axis
figure.autofmt_xdate()
axis.plot(index,stock_open)
```

## Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
```

```python
df_tesla.resample(rule='A').max()
```

```python
### RULE A MEANS YEAR END FREQUENCY
df_tesla.resample(rule='A').max()['Open'].plot()
```

```python
### QUATERLY start frequecny
df_tesla.resample(rule='QS').max()['High'].plot()
```

```python
### Bussiness end frequecny
df_tesla.resample(rule='BA').max()
```

```python
### Bussiness Quaters frequecny
df_tesla.resample(rule='BQS').max()
```

```python
##plotting
df_tesla['Open'].resample(rule='A').mean().plot(kind='bar')
```

```python
df_tesla['Open'].resample(rule='M').max().plot(kind='bar',figsize = (15,6))
```
